```python
import pandas as pd
import matplotlib.pyplot as mat
import seaborn as sns
```

```python
DataFrame = pd.read_csv('adult.csv')
DataFrame.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'Target']
DataFrame
```

## **HISTOGRAM**

```python
col = ['age', 'education-num', 'hours-per-week']
DataFrame[col].hist(bins=10, figsize=(10, 6))
mat.tight_layout()
mat.show()
```

## **DOT**

```python
mat.plot(DataFrame['fnlwgt'], 'ro')
```

```python
col = ['age', 'education-num', 'hours-per-week']
for c in col:
  DataFrame[c].value_counts().plot(kind='bar')
  mat.show()
```

## **LINE**

```python
col = ['age', 'education-num', 'hours-per-week']
for c in col:
  mat.plot(DataFrame[c])
  mat.show()
```

## **SCATTER**

```python
col = ['education-num', 'hours-per-week']
for c in col:
  sns.scatterplot(data=DataFrame, x=col[0], y=col[1])
  mat.show()
```
