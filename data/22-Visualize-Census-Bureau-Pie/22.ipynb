```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mat
import seaborn as sns
```

```python
DataFrame = pd.read_csv('adult.csv')
DataFrame.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'Target']
DataFrame
```

```python
col=['age','education-num','hours-per-week']
DataFrame[col].hist(bins=10,figsize=(10,6))
mat.tight_layout()
mat.show()
```

```python
col = ['count', 'Time']
cnt = [240, 120]
mat.pie(cnt, labels=col)
mat.show()
```

```python
sns.boxplot(data=DataFrame, x='hours-per-week', y='education-num')
mat.plot()
```

## **SCATTER**

```python
sns.scatterplot(data=DataFrame, x='education-num', y='capital-loss')
mat.show()
```

## **BAR**

```python
sns.barplot(data=DataFrame, x='education-num', y='capital-loss')
mat.show()
```

```python
col = ['education-num', 'age']
for i in col:
  DataFrame[i].value_counts().plot(kind='bar')
  mat.show()
```
