```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Cleavland.csv', header=None)

# adding columns
data.columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']

data.head()
```

```python
data.shape
```

## **Data Cleaning**

```python
data.isnull().sum().sum()
```

```python
(data.values == '?').sum()
```

```python
data = data.replace('?', np.nan)
data = data.dropna()
```

```python
data.shape
```

# **Plotting the graphs**

## **1. Histogram**
**Objective: Visualize the distribution of a continuous variable such as age or cholestrol levels in the dataset**

```python
plt.figure(figsize=(10, 8))
plt.hist(data['age'], bins=10, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.title('Age Distribution', size=20)
plt.show()
```

## **2. Dot Plots**
**Objective: Compare the distribution of a categorical variable such as chest pain type or presence of fasting blood sugar, with respect to another variable**

```python
plt.figure(figsize=(12, 9))
sns.stripplot(x='cp', y='trestbps', data = data, jitter=True)
plt.xlabel('Chest Pain Type')
plt.ylabel('Resting Blood Sugar')
plt.title('Chest Pain vs. Resting Blood Sugar', size=20)
plt.legend()
plt.show()
```

## **3. Bar Plots**
**Objective: Compare the values of a categorical variable such as the frequency of different types of heart disease or the distribution of sexes**

```python
plt.figure(figsize=(12, 10))
data['num'].value_counts().plot(kind='bar', color='red')
plt.xlabel('Heart Diseases')
plt.ylabel('Frequency')
plt.title('Heart Disease Distribution', size=20)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
data['sex'].value_counts().plot(kind='bar', color='pink')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.title('Sex Distribution', size=20)
plt.show()
```

## **4. Line Charts**
**Objective: Visualize trends or patterns in a continuous variable such as heart rate (thalach) over time**

```python
plt.figure(figsize=(12, 10))
plt.plot(data['thalach'], color='purple')
plt.xlabel('Index')
plt.ylabel('Heart Rate')
plt.title('Heart Rate Over Time', size=20)
plt.show()
```

## **5. Multivariate analysis using Scatter Plots**
**Objective: Identify relationships or correlations between two continuous variables such as age and cholestrol levels and explore any potential patterns**

```python
plt.figure(figsize=(12, 10))
sns.scatterplot(x='age', y='chol', data=data, hue='num')
plt.xlabel('Age')
plt.ylabel('Cholestrol')
plt.title('Age vs. Cholestrol', size=20)
plt.show()
```
